### Master Image-Label Mapping Table

```python
import os
import xml.etree.ElementTree as ET
import pandas as pd

# 1. Define where your files are located
# (Adjust these paths if your folders are located somewhere else)
xml_dir = "../Dataset/Annotations" 
output_dir = "../Outputs"

# Ensure your Output folder exists before saving files into it
os.makedirs(output_dir, exist_ok=True)

# 2. Create an empty list to store the data row-by-row
all_rows = []

# 3. Loop through every single file in your XML directory
print("Parsing XML files... Please wait.")
for file_name in os.listdir(xml_dir):
    # Only process files that end with .xml
    if file_name.endswith(".xml"):
        xml_path = os.path.join(xml_dir, file_name)
        
        # Parse the XML file tree structure
        tree = ET.parse(xml_path)
        root = tree.getroot()
        
        # Locate the <object> element
        obj = root.find("object")
        if obj is not None:
            # Extract the class label (0 or 1)
            label = obj.find("name").text
            
            # Locate the bounding box coordinates
            bndbox = obj.find("bndbox")
            xmin = int(bndbox.find("xmin").text)
            ymin = int(bndbox.find("ymin").text)
            xmax = int(bndbox.find("xmax").text)
            ymax = int(bndbox.find("ymax").text)
            
            # Determine the corresponding image file name
            image_name = file_name.replace(".xml", ".jpg")
            
            # Combine everything into a single row list
            row_data = [image_name, label, xmin, ymin, xmax, ymax]
            all_rows.append(row_data)

# 4. Convert the gathered list into a structured Pandas DataFrame
df = pd.DataFrame(
    all_rows, 
    columns=["image_name", "label", "xmin", "ymin", "xmax", "ymax"]
)

# 5. Save the DataFrame as a CSV spreadsheet
csv_save_path = os.path.join(output_dir, "image_label_mapping.csv")
df.to_csv(csv_save_path, index=False)

print(f"Success! Master mapping table created and saved to: {csv_save_path}")
print(f"Total entries processed: {len(df)}")
```
